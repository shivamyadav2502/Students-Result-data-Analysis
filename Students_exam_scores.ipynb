```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Drop Unnamed Column

```python
df = df.drop("EthnicGroup", axis=1)
print(df.head())
```

# change weekly study hours column

```python
df["WklyStudyHours"] =df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# gender distrubution

```python
plt.figure(figure= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# from the above chat we have analysed that the number of  females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figure= (5,5))
sns.heatmap(gb, annot = True)
plt.show()
```

# from the abovel chart we have concluded that the education of the parents have a good impact on their scores

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figure= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Realtionship between Parent's Education and Student's Scrore")
plt.show()
```

# from the above chart we have concluded that there is no/negligible import on the student's score due to their parent martial status

# Distribution Of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
print(groupA)
```

```python
import matplotlib.pyplot as plt


groupA = df.loc[df['EthnicGroup'] == "group A"].shape[0]
groupB = df.loc[df['EthnicGroup'] == "group B"].shape[0]
groupC = df.loc[df['EthnicGroup'] == "group C"].shape[0]
groupD = df.loc[df['EthnicGroup'] == "group D"].shape[0]
groupE = df.loc[df['EthnicGroup'] == "group E"].shape[0]

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA, groupB, groupC, groupD, groupE]

print(mlist)
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title("Distribution Of Ethnic Groups")
plt.show()
```
